```python
import numpy as np
```

```python
def read_input(infile):
    ny, nx = 0, 0
    rollmap = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            l = [1 if c == '@' else 0 for c in line.strip()]
            rollmap.append(l)
            ny = len(line)
            nx += 1

    rmap = np.zeros((ny+2, nx+2), dtype=int)
    rmap[1:-2, 1:-1] = np.array(rollmap, dtype=int)
    return rmap

def accessible(rollmap):

    n = 0
    ny, nx = rollmap.shape

    for y in range(1, ny-1):
        for x in range(1, nx-1):
            if rollmap[y, x] == 1 and np.sum(rollmap[y-1:y+2, x-1:x+2]) <= 4:
                n += 1

    return n    

def access_and_remove(rollmap):

    ny, nx = rollmap.shape

    nmap = rollmap.copy()
    n_removed = 1
    n_tot = 0

    while n_removed > 0:
        n_removed = 0
        for y in range(1, ny-1):
            for x in range(1, nx-1):
                if rollmap[y, x] == 1 and np.sum(rollmap[y-1:y+2, x-1:x+2]) <= 4:
                    nmap[y, x] = 0
                    n_removed += 1
                    n_tot += 1
        rollmap = nmap
    return n_tot
    
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = accessible(read_input('input4a.txt'))

print(f'# accessible is {res}')

assert res == 13

print('\nPuzzle case\n-----------\n')

res = accessible(read_input('input4.txt'))

print(f'# accessible is {res}')

assert res == 1384

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = access_and_remove(read_input('input4a.txt'))

print(f'# removed is {res}')

assert res == 43

print('\nPuzzle case\n-----------\n')

res = access_and_remove(read_input('input4.txt'))

print(f'# removed is {res}')

assert res == 8013
```
