```python
def read_input(infile):
    banks = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            banks.append([int(i) for i in line.strip()])
    return banks

def find_joltage(banks):
    maxval = 0

    for bank in banks:
        val = 0
        for i in range(9, 0, -1):
            try:
                idx = bank.index(i)
                if idx == len(bank):
                    val = max(val, i)
                    continue
                val = max(val, i*10 + max(bank[idx+1:]))
            except ValueError:
                continue
        maxval += val

    return maxval

def find_biggest_number(l, bank):

    if l == 1:
        return max(bank)

    mval = max(bank[:-(l-1)])
    idx = bank.index(mval)
    return mval * 10**(l-1) + find_biggest_number(l-1, bank[idx+1:])


def find_more_joltage(banks):
    maxval = 0

    for bank in banks:
        maxval += find_biggest_number(12, bank)

    return maxval
        
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = find_joltage(read_input('input3a.txt'))

print(f'Joltage is {res}')

assert res == 357

print('\nPuzzle case\n-----------\n')

res = find_joltage(read_input('input3.txt'))

print(f'Joltage is {res}')

assert res == 17278

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = find_more_joltage(read_input('input3a.txt'))

print(f'Joltage is {res}')

assert res == 3121910778619

print('\nPuzzle case\n-----------\n')

res = find_more_joltage(read_input('input3.txt'))

print(f'Joltage is {res}')

assert res == 171528556468625
```
