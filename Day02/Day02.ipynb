```python
def read_input(infile):
    groups = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            grps = line.strip().split(',')
            for g in grps:
                groups.append([int(i) for i in g.split('-')])

    return groups

def find_bad_ids(groups):

    n_bad = 0
    for grp in groups:
        for cid in range(grp[0], grp[1]+1):
            s = str(cid)
            l = len(s)
            if l % 2 != 0:
                continue
            h1 = l//2
            h2 = l//2
            if s[:h1] == s[h2:]:
                n_bad += cid

    return n_bad

def find_bad_ids2(groups):

    bad_ids = []
    n_bad = 0
    for grp in groups:
        for cid in range(grp[0], grp[1]+1):
            s = str(cid)
            l = len(s)

            for ll in range(1, l//2+1):
                if l % ll != 0:
                    continue
                n = l // ll
                if s == n*s[:ll] and cid not in bad_ids:
                    bad_ids.append(cid)

    return sum(bad_ids)
```

```python
find_bad_ids2(read_input('input2.txt'))
```
