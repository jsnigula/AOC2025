```python
def read_input(infile):
    instructions = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            l = line.strip()
            if l[0] == 'L':
                instructions.append(int(l[1:]) * -1)
            else:
                instructions.append(int(l[1:]))
    return instructions

def get_password(instructions):
    val = 50
    password = 0
    for i in instructions:
        val = (val + i) % 100
        if val == 0:
            password += 1

    return password

def get_password2(instructions):
    val = 50
    password = 0
    for i in instructions:
        nval = (val + i)
        if nval == 0:
            password += 1
        if nval > 0:
            password += nval // 100
        elif nval < 0:
            password += abs(nval // -100)#  + 1
            if val > 0:
                password += 1
        val = nval % 100
    return password
```

```python
get_password2(read_input('input1a.txt'))
```
